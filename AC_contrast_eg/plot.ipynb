```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator

data = scipy.io.loadmat('AC.mat')
usol = data['uu']
t_star = data['tt'][0]
x_star = data['x'][0]
TT, XX = np.meshgrid(t_star, x_star)

plt.figure(1)
plt.pcolor(TT, XX, usol, cmap='jet', vmin=-1, vmax=1)
cbar = plt.colorbar(format='%1.1f')
cbar.ax.tick_params(labelsize=15)
plt.xlabel('$t$', fontsize=18)
plt.ylabel('$x$', fontsize=18)
plt.xticks(size=15)
plt.yticks(size=15)
plt.title(r'Exact solution $u(t,x)$', fontsize=22)
plt.show()

data = np.load('AC_pred_pinn.npy')
plt.figure(2)
plt.pcolor(TT, XX, data, cmap='jet', vmin=-1, vmax=1)
cbar = plt.colorbar(format='%1.1f')
cbar.ax.tick_params(labelsize=15)
plt.xlabel('$t$', fontsize=18)
plt.ylabel('$x$', fontsize=18)
plt.xticks(size=15)
plt.yticks(size=15)
plt.title(r'PINN prediction $u(t,x)$', fontsize=22)
plt.show()

data = np.load('AC_pred_our.npy')
plt.figure(3)
plt.pcolor(TT, XX, data, cmap='jet', vmin=-1, vmax=1)
cbar = plt.colorbar(format='%1.1f')
cbar.ax.tick_params(labelsize=15)
plt.xlabel('$t$', fontsize=18)
plt.ylabel('$x$', fontsize=18)
plt.xticks(size=15)
plt.yticks(size=15)
plt.title(r'TL-DPINN prediction $u(t,x)$', fontsize=22)
plt.show()

data = np.load('AC_residual_pinn.npy')
plt.figure(4)
plt.plot(np.linspace(0,1,len(data)-2),data[2:],'d-')
plt.xlabel('$t$', fontsize=18)
plt.ylabel(r'$\mathcal{L}(t_i,\theta)$', fontsize=18)
plt.xticks(size=15)
plt.yticks(size=15)
plt.yscale('log')
plt.show()

data = np.load('AC_residual_our.npy')
plt.figure(5)
plt.plot(np.linspace(0,1,len(data)),data,'d-')
plt.xlabel('$t$', fontsize=18)
plt.ylabel(r'$\mathcal{L}(t_i,\theta)$', fontsize=18)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```
